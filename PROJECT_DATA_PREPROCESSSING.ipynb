```python
# prompt: write a funcvtion and apply on df['price ] to convert all values on crore and name of columns price in crore and all values upto 2 decible places  and drop previous price col

def convert_to_crore(price_str):
    try:
        price_str = str(price_str)
        if 'Cr' in price_str:
            price_in_cr = float(price_str.split(' ')[0])
            return round(price_in_cr, 2)
        elif 'L' in price_str:
            price_in_l = float(price_str.split(' ')[0])
            price_in_cr = price_in_l / 100
            return round(price_in_cr, 2)
        else:
            price_val = float(price_str)
            price_in_cr = price_val / 10000000
            return round(price_in_cr, 2)
    except:
        return np.nan

df['price_in_crore'] = df['price'].apply(convert_to_crore)
df.drop('price', axis=1, inplace=True)
```
